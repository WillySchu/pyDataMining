```python
import numpy as np
import pandas as pd
```

```python
from fbprophet import Prophet
```

```python
data = pd.read_csv('sessionsbysource.tsv', sep='\t', parse_dates=[1])
```

```python
data = data.groupby(['ga:source', 'ga:date']).sum()
data['y'] = data['ga:sessions']
del data['ga:sessions']
```

```python
direct = data.ix['(direct)']
```

```python
direct['ds'] = direct.index
```

```python
m = Prophet()
```

```python
fit = m.fit(direct)
```

```python
future = m.make_future_dataframe(periods=10)
```

```python
fcast = m.predict(future)
```

```python
fcast
```

```python
m.plot(fcast)
```

```python
m.plot_components(fcast)
```
